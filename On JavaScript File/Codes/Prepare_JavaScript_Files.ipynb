```python
def delete_duplecte(df):
  print(df.shape)
  to_delete = []
  for i in tqdm(range(df.shape[0])):
    for j in range(df.shape[0]):
      if df['ScriptPath'][i] == df['ScriptPath'][j] and i!=j:
        to_delete.append(i)
  df = df.drop(labels=to_delete, axis=0)
  df = df.mask(df.eq(np.nan)).dropna()
  print(df.shape)
  return df
```

```python
def creat_data(df_1,df_2,b=None): # (df_1=df_0,df_2=df_1,b=None)
  if b is None:
    if df_1.shape[0]>df_2.shape[0]:
      df_1 = df_1.sample(df_2.shape[0])
    else:
      df_2 = df_2.sample(df_1.shape[0])
  else:
    if df_1.shape[0]>df_2.shape[0]:
      df_1 = df_1.sample(int(df_2.shape[0]*b))
    else:
      df_2 = df_2.sample(int(df_1.shape[0]*b))
    frames = [df_1, df_2]
    result = pd.concat(frames)
    df = shuffle(result)
    return df
```

```python
def save_object(obj, filename,path):
    """
    _ INPUT (obj) THE OBJECT WE NEED SAVW IT (filename) THE NAME OF OBJECT
    """
    filename = os.path.join(path,filename)
    with open(filename+".pkl", 'wb') as outp:
        pickle.dump(obj, outp, pickle.HIGHEST_PROTOCOL)
    outp.close()
def load_object(filename,path):
    """
    _ INPUT THE NAME OF OBJECT WE NEED LOAD IT
    """
    filename = os.path.join(path,filename)
    with open(filename+".pkl", 'rb') as outp:
        loaded_object = pickle.load(outp)
    outp.close()
    return loaded_object
```
